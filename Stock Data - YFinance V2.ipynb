```python
import yfinance as yf
import pandas as pd
from datetime import datetime

def convert_date_format(date_str):
    # Parse the input date string
    date = datetime.strptime(date_str, '%Y%m%d')

    # Convert the date to the desired format
    formatted_date = date.strftime('%Y-%m-%d')

    return formatted_date

# Read ticker symbols and company names from CSV file
tickers_df = pd.read_csv('bank_master_list.csv')
tickers = tickers_df['Ticker'].tolist()
company_names = tickers_df['Company Name'].tolist()

# Create a space-separated string of tickers
ticker_list = ' '.join(tickers)

# Read quarter end dates from "quarter_end_dates.csv"
quarter_dates_df = pd.read_csv('quarter_end_dates.csv', dtype={'quarter_end_date': str})
dates_list = quarter_dates_df['quarter_end_date'].tolist()
quarter_ids = quarter_dates_df['quarter_id'].tolist()

combined_df = pd.DataFrame()  # Initialize an empty DataFrame for combining the data

# Iterate through the dates_list
for s_date, quarter_id in zip(dates_list, quarter_ids):
    c_date = convert_date_format(s_date)
    print(s_date, c_date)
    # Iterate backward to find a valid trading day with data
    start_date = pd.to_datetime(c_date)
    end_date = start_date + pd.DateOffset(days=1)
    while True:
        # Download data for the specific date range
        df = yf.download(tickers=ticker_list, start=start_date, end=end_date, group_by='Ticker')

        # Check if the DataFrame is empty
        if not df.empty:
            break

        # Decrement the current date by one day
        start_date -= pd.DateOffset(days=1)

    df = df.stack(level=0).rename_axis(['Date', 'Ticker']).reset_index(level=1)
    df['quarter_id'] = quarter_id  # Add the quarter_id column
    combined_df = pd.concat([combined_df, df], axis=0)  # Concatenate the current DataFrame with the combined DataFrame

# Save combined DataFrame to CSV
combined_df.to_csv('quarterly_stock_data.csv', index=True)
print(combined_df)
print("Done!")
```
